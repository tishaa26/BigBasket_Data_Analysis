```python
#MAXIMUM MARKET PRICE
print("Maximum Market price:", df['market_price'].max())
```

```python
#COUNTS OF CATEGORY AND SUBCATEGORY
category_counts = df['category'].value_counts()
sub_category_counts = df['sub_category'].value_counts()

print("Category Counts:\n", category_counts)
print("\nSub-category Counts:\n", sub_category_counts)
```

```python
#DISTRIBUTION OF SALE PRICE
plt.hist(df['sale_price'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.title('Distribution of Sale Price')
plt.show()
```

```python
#DISTRIBUTION OF RATINGS
plt.figure(figsize=(8, 6))
plt.hist(df['rating'], bins=10, color='orange', edgecolor='black')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#SALES VS MARKET PRICE
plt.figure(figsize=(8, 6))
plt.scatter(df['market_price'], df['sale_price'], color='purple', alpha=0.7)
plt.title('Market Price vs Sale Price')
plt.xlabel('Market Price')
plt.ylabel('Sale Price')
plt.grid(True)
plt.show()
```

```python
#BOX PLOT OF SALES PRICE CATEGORY
plt.figure(figsize=(10, 6))
boxplot = df.boxplot(column='sale_price', by='category', vert=False, patch_artist=True, showfliers=False)

# Customizing boxplot colors
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(boxplot['boxes'], colors):
    patch.set_facecolor(color)

plt.xlabel('Sale Price')
plt.ylabel('Category')
plt.title('Box Plot of Sale Price by Category')
plt.grid(True)
plt.show()
```

```python
#NUMBER OF UNIQUE BRANDS
unique_brands = df['brand'].nunique()
print("Number of Unique Brands:", unique_brands)
```

```python
#AVERAGE SALE BY PRICE
avg_sale_price_by_brand = df.groupby('brand')['sale_price'].mean()
print("Average Sale Price by Brand:\n", avg_sale_price_by_brand)
```

```python
#COUNTS OF PRODUCT IN EACH CATEGORY
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='category', palette='pastel')
plt.title('Count of Products in Each Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#MEAN SALE BY CATEGORY
category_mean_sale_price = df.groupby('category')['sale_price'].mean()
print("Mean Sale Price by Category:\n", category_mean_sale_price)
```

```python
#CALCULATING TOTAL BY SALES PRICE BY CATEGORY
total_sale_price_by_category = {category: np.sum(df[df['category'] == category]['sale_price'])
                                for category in df['category'].unique()}

print("Total Sale Price by Category:")
for category, total_sale_price in total_sale_price_by_category.items():
    print(category, ":", total_sale_price)
```

```python
#CALCULATING THE COUNT OF PRODUCTS BY CATEGORY
count_by_category = {category: np.sum(df['category'] == category)
                     for category in df['category'].unique()}

print("Count of Products by Category:")
for category, count in count_by_category.items():
    print(category, ":", count)
```

```python
#CALCULATING SALES PRICE STATISTICS USING GROUP BY CATEGORY
sale_price_stats_by_category = df.groupby('category')['sale_price'].agg(['mean', 'median', 'std', 'min', 'max'])
print("Sale Price Statistics by Category:")
print(sale_price_stats_by_category)
```

```python
#ADDING A NEW COLUMN FOR PRICE DIFFERENCE
df['price_difference'] = df['market_price'] - df['sale_price']
print("Dataset with Price Difference Column:")
df.head()
```

```python
#ADDING A NEW COLUMN FOR DISCOUNT PERCENTAGE
df['discount_percentage'] = ((df['market_price'] - df['sale_price']) / df['market_price']) * 100

# Filter products with discount
discounted_products = df[df['discount_percentage'] > 0]

print("Products with Discount Percentage:")
discounted_products[['index', 'product', 'discount_percentage']].head()
```

```python
#CALCULATING THE TOTAL REVENUE BY BRAND
total_revenue_by_brand = df.groupby('brand')['sale_price'].sum()
print("Total Revenue by Brand:")
print(total_revenue_by_brand)
```
